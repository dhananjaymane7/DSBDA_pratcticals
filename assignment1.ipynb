```python
# //Scatter Plot
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("psd_coffee.csv")
plt.scatter(data['Year'],data['Imports'])
plt.title("Scatter plot")
plt.xlabel('Year')
plt.ylabel('Imports')
plt.show()
```

```python

# //Color Bar
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("psd_coffee.csv")
plt.scatter(data['Year'],data['Imports'], c=data['Total Distribution'],s=data['Total Supply'])
plt.title("Scatter plot")
plt.xlabel('Year')
plt.ylabel('Imports')
plt.colorbar()
plt.show()
```
