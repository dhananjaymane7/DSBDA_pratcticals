## iris dataset


```python
iris = pd.read_csv('iris.csv')
```

```python
iris.describe()
```

```python
iris.info()
```

```python
meandf = iris.groupby(['Species'], as_index = False).agg('mean')
```

```python
meandf
```

```python
mediandf = iris.groupby(['Species'], as_index = False).agg('median')
```

```python
mediandf
```

```python
stddf = iris.groupby(['Species'], as_index = False).agg('std')
```

```python
stddf
```

```python
stddf.plot('Species',['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],figsize=(15, 8))
```
