```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
iris=load_iris()

data=pd.DataFrame(iris.data, columns=iris.feature_names)
data['label']=iris.target
data.sample(5)
```

```python
data.info()
```

```python
#Histogram
sns.histplot(data=data, x='sepal length (cm)',kde=True)
```

```python
sns.histplot(data=data, x='sepal width (cm)', kde=True, color="orange")
```

```python
sns.histplot(data=data, x='petal length (cm)', kde=True, color="green")
```

```python
#Boxplots

figure=plt.figure(figsize=(12,8))  #12 inches wide and 8 inches tall
sns.boxplot(data=data)
plt.show()
```

```python
from matplotlib.cbook import boxplot_stats
stats=boxplot_stats(data['sepal width (cm)'])
stats
```

```python
outliers=stats[0].get("fliers")
outliers
```

```python
sepal_width = data['sepal width (cm)']

# Get the outliers from the 'stats' dictionary
outliers = stats[0].get("fliers")

# Remove outliers from the 'sepal_width' array
cleaned_sepal_width = sepal_width[~np.isin(sepal_width, outliers)]

# Show the cleaned sepal width data
print(cleaned_sepal_width)
```
